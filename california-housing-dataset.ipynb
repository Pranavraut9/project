```python
def split_train_test(data,test_ratio):
    #shuffled_indices.random.seed(42)
    shuffled_indices=np.random.permutation(len(data))
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled_indices[:test_set_size]
    train_indices=shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
train_set,test_set=split_train_test(cf,0.2)
```

```python
print(len(train_set))
print(len(test_set))
```

```python
from zlib import crc32
def test_set_check(identifier,test_ratio):
    return crc32(np.int64(identifier))&0xffffffff< test_ratio * 2**32
```

```python
def split_train_test_by_id(data,test_ratio,id_column):
    ids=data[id_column]
    in_test_set=ids.apply(lambda id_:test_set_check(id_,test_ratio))
    return data.loc[~in_test_set],data.loc[in_test_set]
```

```python
cf_with_id=cf.reset_index()
train_set,test_set=split_train_test_by_id(cf_with_id,0.2,'index')
print(train_set.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scalar',StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs=list(housing_num)
cat_attribs=['ocean_proximity']
```

```python
full_pipeline=ColumnTransformer([
    ('num',num_pipeline,num_attribs),
    ('cat',OneHotEncoder(),cat_attribs)
])
```

```python
housing_prepared=full_pipeline.fit_transform(x)
```

```python
from flask import Flask
from flask import render_template
from flask import request
import joblib


app= Flask(__name__)
model= joblib.load('my_cal_model.pkl')
@app.route('/')
def hello(name=None):
    return render_template('index.html',name=name)

@app.route('/result',methods=['GET','POST'])
def get_value():
    if request.method=='POST':
        long = float(request.form.get('long'))
        lat = float(request.form.get('lat'))
        hma = float(request.form.get('hma'))
        tr = float(request.form.get('tr'))
        tbr = float(request.form.get('tbr'))
        popu = float(request.form.get('popu'))
        hohold = float(request.form.get('hohold'))
        minc = float(request.form.get('minc'))
        ocean = float(request.form.get('ocean'))
        
        predict_value=model.predict([[long,lat,hma,tr,tbr,popu,hohold,minc,ocean,0,0,0,0]]) 
        print(predict_value[0])

    return render_template('result.html',prediction=[predict_value[0],long,lat])
```
